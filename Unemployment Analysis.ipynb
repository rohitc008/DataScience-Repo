## Data Visualization For Dataset 1

```python
df[' Frequency'] = df[' Frequency'].map({" Monthly":"/Month",
                                            "Monthly":"/Month"})
```

```python
df[' Frequency'].value_counts()
```

```python
plt.figure(figsize=(15, 10))
for i,z in enumerate([' Estimated Unemployment Rate (%)',' Estimated Employed', ' Estimated Labour Participation Rate (%)']):
    plt.subplot(2,2,i+1)
    sns.histplot(x=z, hue="Area", data=df, kde=True, palette="Set1")
    
    plt.ylabel("Count")
plt.title("Histogram of Estimated Employment Rate by Area")
plt.show()
```

```python
num_data= df.select_dtypes(include='number')

plt.style.use("ggplot")
plt.figure(figsize=(10,10))
sns.heatmap(num_data.corr(),annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
fig = plt.figure(figsize = (8, 6))
sns.lineplot(y=' Estimated Unemployment Rate (%)', x=' Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
y=df2[' Estimated Unemployment Rate (%)']
x=df2['Region']
plt_1 = plt.figure()
plt.title('Umemployment Rate')
plt.xlabel("States")
plt.ylabel("Estimated Unemployment rate",)
plt.xticks(rotation='vertical',fontsize=12)
sns.histplot(x,)
```

```python
fig = plt.figure()
plt.scatter(df2['Region.1'], df2[' Estimated Labour Participation Rate (%)'])
plt.title('Labour Participation according to Region')
plt.xlabel('Region')
plt.ylabel('Labour Participation Rate')
plt.show()
```

```python
plt.figure(figsize=(12, 12))
df2.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=df2)
plt.show()
```
